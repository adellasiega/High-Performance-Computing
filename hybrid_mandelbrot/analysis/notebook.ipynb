```python
# define a function to plot the data in csv in ../results
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def plot_data(filename, x, y, x_label, y_label, title):
    data = pd.read_csv(filename)
    x = data[x]
    y = data[y]
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_ylim(bottom=0)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
    plt.show()

plot_data('../results/mpi_strong_scaling.csv', 'mpi_size', 'time_seconds', 'mpi_size', 'time_seconds', 'MPI Strong Scaling')
plot_data('../results/mpi_weak_scaling.csv', 'mpi_size', 'time_seconds', 'mpi_size', 'time_seconds', 'MPI Weak Scaling')
plot_data('../results/omp_strong_scaling.csv', 'omp_threads', 'time_seconds', 'omp_threads', 'time_seconds', 'OMP Strong Scaling')
plot_data('../results/omp_weak_scaling.csv', 'omp_threads', 'time_seconds', 'omp_threads', 'time_seconds', 'OMP Weak Scaling')


```
