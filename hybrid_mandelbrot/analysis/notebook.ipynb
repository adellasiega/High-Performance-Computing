```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# MPI strong scaling

```python
# Plot the data in ../results/mpi_strong_scaling.csv

data = pd.read_csv('../results/mpi_strong_scaling.csv')
mpi_size = data['mpi_size']
time_seconds = data['time_seconds']
fig, ax = plt.subplots()
ax.plot(mpi_size, time_seconds)
#add Ahmdal's law line in a dotted line
ax.plot(mpi_size, time_seconds[0] / mpi_size, linestyle='dotted')
ax.legend(['Measured', 'Ideal'])
ax.set_ylim(bottom=0)
ax.set_xlabel('Number of processes')
ax.set_ylabel('Time (s)')
ax.set_title('MPI Strong Scaling')
ax.grid()
plt.savefig('../images/mpi_strong_scaling.png')
plt.show()
```

```python
# MPI strong scaling speedup
#Speedup mpi weak scaling
fig, ax = plt.subplots()
ax.plot(mpi_size, time_seconds[0]/time_seconds)
#add ahmdal's law line
ax.plot(mpi_size, mpi_size, linestyle='dotted')
ax.set_xlabel('Number of processes')
ax.legend(['Measured', 'Ideal'])
ax.set_ylabel('Speedup')
plt.title('MPI Strong Scaling Speedup')
ax.grid()
plt.savefig('../images/mpi_strong_scaling_speedup.png')
plt.show()
```

# OpenMP strong scaling

```python
# OpenMP strong scaling
data = pd.read_csv('../results/omp_strong_scaling.csv')
threads = data['omp_threads']
time_seconds = data['time_seconds']
fig, ax = plt.subplots()
ax.plot(threads, time_seconds)
#add Ahmdal's law line in a dotted line
ax.plot(threads, time_seconds[0] / threads, linestyle='dotted')
ax.legend(['Measured', 'Ahmdal\'s law'])
ax.set_ylim(bottom=0)
ax.set_xlabel('Number of threads')
ax.set_ylabel('Time (s)')
ax.set_title('OpenMP Strong Scaling')
ax.grid()
plt.savefig('../images/omp_strong_scaling.png')
plt.show()
```

```python
# OpenMP strong scaling speedup
fig, ax = plt.subplots()
ax.plot(threads, time_seconds[0]/time_seconds, label='Measured')
#add Ahmdal's law line in a dotted line
ax.plot(threads, threads, linestyle='dotted', label='Ideal')
ax.set_xlabel('Number of threads')
ax.set_ylabel('Speedup')
plt.title('OpenMP Strong Scaling Speedup')
ax.grid()
plt.legend()
plt.savefig('../images/omp_strong_scaling_speedup.png')
plt.show()

```
